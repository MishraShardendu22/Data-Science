`matplotlib.pyplot.figure()` lets you control both the **size** and the **resolution** of your plots.

* **figsize = (width, height)**

  * Defines the width and height of the figure in **inches**.
  * Example: `figsize=(6, 4)` → 6 inches wide and 4 inches tall.

* **dpi = resolution**

  * DPI (dots per inch) sets how many pixels fit in each inch.
  * Higher DPI = more pixels per inch = sharper and more detailed image.
  * Example: `dpi=100` → 100 pixels per inch.

* **Pixel dimensions = width × dpi , height × dpi**

  * Example: `figsize=(6,4), dpi=100` → figure size = **600 × 400 pixels**.

# Matplotlib Figure and Axes Placement

## Figure

```python
fig = plt.figure(figsize=(20, 8), dpi=100)
```

* **figsize**: `(width, height)` in **inches**.
* **dpi**: dots per inch → resolution.
* Pixel size = `width × dpi` by `height × dpi`.
* Here: `20 × 100 = 2000 px` wide, `8 × 100 = 800 px` tall.

---

## Axes with `add_axes()`

Syntax:

```python
ax = fig.add_axes([left, bottom, width, height])
```

* Values are fractions of the **figure dimensions** (0 to 1).
* `left`: distance from left edge.
* `bottom`: distance from bottom edge.
* `width`: width of axis (fraction of figure).
* `height`: height of axis (fraction of figure).

---

## Example

```python
axis1 = fig.add_axes([0.1, 0.1, 0.4, 0.8])  # Main plot
axis1.plot(x, y)

axis2 = fig.add_axes([0.3, 0.4, 0.3, 0.3])  # Inset plot
axis2.plot(x, y)
```

* **axis1**: Starts 10% from left, 10% from bottom, covers 40% width, 80% height.
* **axis2**: Inset starting 30% from left, 40% from bottom, covering 30% width and height.


### 1. **General Plots (Matplotlib)**

* **Scatter Plot**: Shows how **two numeric variables relate**. Each point = one data point. Focused on correlation/trends/outliers.
  Example: `plt.scatter(x, y)`

* **Histogram**: Shows **frequency of values** for a single numeric variable. Groups values into bins.
  Example: `plt.hist(data, bins=10)`

* **Box Plot**: Shows **spread, median, quartiles, outliers**. Good for comparing groups.
  Example: `plt.boxplot(data)`

* **Saving Plot**: Not a plot type—just exporting any plot to a file:
  `plt.savefig("plot.png")`

**Key:** These are general-purpose plots, not limited to any library. Focus is either relationships or distribution summaries.

---

### 2. **Distribution Plots (Seaborn)**

Focus: **Understanding distributions and relationships in one or two variables**.

* **Histplot / Distplot**: Frequency distribution of one variable. `histplot` is the modern version.
  Use: Check distribution shape, skewness, modality.

* **Jointplot**: Combines scatter plot **(x vs y)** + marginal histograms/KDEs for each variable.
  Use: See correlation **and** individual distributions together.

* **Pairplot**: Grid of scatter plots for all pairs of numeric variables + histograms on the diagonal.
  Use: Explore relationships across the dataset quickly.

* **Rugplot**: Tiny lines along axes for each data point. Usually combined with histograms/KDEs.
  Use: Show actual data density visually.

**Key:** Distribution plots are mostly exploratory for one or two numeric variables, often with density info.

---

### 3. **Categorical Data Plots (Seaborn)**

Focus: **Data grouped by categories**.

* **Boxplot / Violinplot**: Show distribution per category. Violin also shows density shape.
* **Stripplot / Swarmplot**: Show individual points per category (swarm avoids overlap).
* **Barplot / Countplot**: Summarize counts or aggregated values per category.

**Key:** These focus on comparing **categories**, not numeric-to-numeric relationships.

---

**Summary of Differences**

| Type              | Main Purpose         | Variables                 | Focus                                 |
| ----------------- | -------------------- | ------------------------- | ------------------------------------- |
| Scatter           | Relation             | 2 numeric                 | Trend/correlation                     |
| Histogram         | Distribution         | 1 numeric                 | Frequency/skew                        |
| Boxplot           | Distribution summary | 1 numeric (+categories)   | Quartiles/outliers                    |
| Jointplot         | Combined             | 2 numeric                 | Correlation + distributions           |
| Pairplot          | Multiple relations   | All numeric               | Dataset-wide patterns                 |
| Rugplot           | Data density         | 1 numeric                 | Shows actual points along axis        |
| Categorical plots | Compare categories   | 1 numeric + 1 categorical | Distribution/aggregation per category |

## KDE = **Kernel Density Estimation**.

It’s a way to estimate the **probability density function (PDF)** of a dataset. Instead of just counts in bins (like a histogram), KDE places a smooth curve over the data.

* Each data point is replaced by a small “bump” (a kernel, usually Gaussian).
* All bumps are added up to make a continuous curve.
* The area under the curve = 1 (like a probability distribution).

**Use:**

* See the *shape* of distribution more smoothly than histograms.
* Detect modes (peaks), skewness, or multimodal patterns.

`lmplot()` is a **Seaborn function to plot a linear regression model** between two numeric variables.

* **Purpose:** Show **relationship/trend** between `x` and `y` and optionally fit a regression line.

* **Features:**

  * Adds a **regression line** with confidence interval.
  * Can handle **categorical separation** using `hue`, `col`, or `row`.
  * Combines scatter + regression in one plot.

* **Example:**

```python
sns.lmplot(x='total_bill', y='tip', data=df)
```

Shows how tips increase with bill amounts, with a regression line predicting the trend.

* **Use case:** Predicting or visualizing trends between two numeric variables and seeing correlation strength.
