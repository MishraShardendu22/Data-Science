# Learning Pandas

**Pandas** comes from Panel Data

**Series in Pandas** A Series is a one-dimensional labeled array capable of holding any data type. The axis labels are collectively called the index.
A Series is a one-dimensional labeled array (like a single column), while a

**DataFrame** is a two-dimensional labeled table composed of one or more Series

**Row** = horizontal (left-right).
**Column** = vertical (up-down).

**Axis** 0 is row.

**INNER JOIN (how='inner')** - Only rows with matching keys in both DataFrames are kept.

**LEFT JOIN (how='left')** - All rows from the left DataFrame are kept. If no match in the right DataFrame, the result has NaN for right-side columns.

**RIGHT JOIN (how='right')** - All rows from the right DataFrame are kept. If no match in the left DataFrame, the result has NaN for left-side columns.

**OUTER JOIN (how='outer')** - All rows from both DataFrames are kept. Missing values are filled with NaN.

```python
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})
employees
```

```python
# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
salaries
```

```python
pd.merge(employees,salaries,on='employee_id',how ='inner')
```

```python
pd.merge(employees,salaries,on='employee_id',how ='outer')
```

```python
pd.merge(employees,salaries,on='employee_id',how ='left')
```

```python
pd.merge(employees,salaries,on='employee_id',how ='right')
```

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}

df = pd.DataFrame(data)
df
```

```python
# Group by Category and calculate the sum of Sales
cat = df.groupby('Category')['Sales'].sum()
cat
```

```python
# Group by Category and Store
cat = df.groupby(['Category','Store'])['Sales'].sum()
cat
```

```python
df['Sales'].mean()
```

```python
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
               'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary',
            'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice']
}

df = pd.DataFrame(data)
df
```

```python
df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
df
```

```python
pd.pivot_table(df,values = 'Sales',index = 'Region',columns = 'Product', aggfunc="mean")
```

```python
pivot2 = pd.pivot_table(df, values=['Sales', 'Units'], index='Region', columns='Product')
pivot2
```

```python
pd.crosstab(df['Region'],df['Product'])
```
